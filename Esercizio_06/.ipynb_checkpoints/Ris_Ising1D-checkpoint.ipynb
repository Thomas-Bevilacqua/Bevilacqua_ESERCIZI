## 1D Ising model
In tutto l'esercizio vengono utilizzate le unità $k_B=\mu_B=1.$

Per prima cosa inserisco due funzioni 'Save_spin' e 'Load_spin' per salvare e caricare una data configurazione (per ricominciare la simulazione). Viene automaticamente salvata la configurazione iniziale.

Successivamente modifico la funzione 'Measure' in modo da poter misurare tutte le grandezze richieste; vengono usate le formule date nel testo dell'esercizio.

Scrivo poi gli algoritmi Metropolis e Gibbs della funzione 'Move' e modifico la funzione 'Averages' in modo da stampare tutto. Riporto qui i risultati 'test' effettuati a $T=2~\mathrm{K}$. Ho posto $h=0$ tranne nel valutare la magnetizzazione per cui vale $h=0.02$. Ho scelto di usare 100 blocks ognuno con $10^5$ MC steps.

La simulazione viene effettuata con i parametri dati in input: per ricavare i risultati per la magnetizzazione, che richiedono $h=0.02$, bisogna dunque modificare il file di input e rifare la simulazione. Naturalmente anche gli altri files dei risultati vengono sovrascritti.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

nsteps = 100000

plt.rc('text', usetex=True)
plt.figure(figsize=(20,20))

plt.suptitle('Metropolis test at $2~\mathrm{K}$',fontsize=20)
plt.subplot(221)
plt.title('Average internal energy (per particle)')
bl, E, errE = np.loadtxt("Ris_2K_(test)/output_ene.dat", usecols=(0,2,3), delimiter='   ', unpack='true')
plt.errorbar(bl*nsteps, E ,yerr=errE)
plt.xlabel('$\#$ of MC steps')
plt.ylabel('$<U>/N$')
plt.grid(True)

plt.subplot(222)
plt.title('Average heat capacity')
bl, c, errc = np.loadtxt("Ris_2K_(test)/output_heat.dat", usecols=(0,2,3), delimiter='   ', unpack='true')
plt.errorbar(bl*nsteps, c, yerr=errc)
plt.xlabel('$\#$ of MC steps')
plt.ylabel('$<C>$')
plt.grid(True)

plt.subplot(223)
plt.title('Average magnetization ($h=0.02$)')
bl, m, errm = np.loadtxt("Ris_2K_(test)/output_mag.dat", usecols=(0,2,3), delimiter='   ', unpack='true')
plt.errorbar(bl*nsteps, m, yerr=errm)
plt.xlabel('$\#$ of MC steps')
plt.ylabel('$<M>$')
plt.grid(True)

plt.subplot(224)
plt.title('Average susceptibility')
bl, x, errx = np.loadtxt("Ris_2K_(test)/output_chi.dat", usecols=(0,2,3), delimiter='   ', unpack='true')
plt.errorbar(bl*nsteps, x, yerr=errx)
plt.xlabel('$\#$ of MC steps')
plt.ylabel('$<\chi>$')
plt.grid(True)

plt.show()
```

Osservando questi risultati potrei equilibrare il sistema effettuando $5\cdot10^6$ MC steps prima di iniziare la simulazione. Tuttavia, dato che nell'effettuare la simulazione a varie temperature il processo inizia a diventare lungo, decido di non equilibrare il sistema ogni volta anche perché le oscillazioni sono contenute in intervalli molto stretti rispetto alle grandezze misurate.

Ripeto la stessa simulazione con l'algoritmo Gibbs.

```python
plt.figure(figsize=(20,20))

plt.suptitle('Gibbs test at $2~\mathrm{K}$',fontsize=20)
plt.subplot(221)
plt.title('Average internal energy (per particle)')
bl, E, errE = np.loadtxt("Ris_2K_(test)/output_eneG.dat", usecols=(0,2,3), delimiter='   ', unpack='true')
plt.errorbar(bl*nsteps, E ,yerr=errE)
plt.xlabel('$\#$ of MC steps')
plt.ylabel('$<U>/N$')
plt.grid(True)

plt.subplot(222)
plt.title('Average heat capacity')
bl, c, errc = np.loadtxt("Ris_2K_(test)/output_heatG.dat", usecols=(0,2,3), delimiter='   ', unpack='true')
plt.errorbar(bl*nsteps, c, yerr=errc)
plt.xlabel('$\#$ of MC steps')
plt.ylabel('$<C>$')
plt.grid(True)

plt.subplot(223)
plt.title('Average magnetization ($h=0.02$)')
bl, m, errm = np.loadtxt("Ris_2K_(test)/output_magG.dat", usecols=(0,2,3), delimiter='   ', unpack='true')
plt.errorbar(bl*nsteps, m, yerr=errm)
plt.xlabel('$\#$ of MC steps')
plt.ylabel('$<M>$')
plt.grid(True)

plt.subplot(224)
plt.title('Average susceptibility')
bl, x, errx = np.loadtxt("Ris_2K_(test)/output_chiG.dat", usecols=(0,2,3), delimiter='   ', unpack='true')
plt.errorbar(bl*nsteps, x, yerr=errx)
plt.xlabel('$\#$ of MC steps')
plt.ylabel('$<\chi>$')
plt.grid(True)

plt.show()
```

Noto che i risultati sono compatibili tranne nel caso della magnetizzazione che viene più piccola rispetto a quella valutata con l'algoritmo Metropolis. In questo caso si potrebbero scartare solo i primi $2\cdot10^6$ MC steps per equilibrare il sistema.

Ora effettuo la simulazione coi due metodi nel range di temperature $[0.5,2]~\mathrm{K}$ a steps di $0.05~\mathrm{K}$.

```python
#Expected functions
points = 100
T = np.linspace(0.4,2.1, num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th

#Internal energy (per particle)
e = -J*( th + ch*thN )/( 1 + thN )

#Heat capacity
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

#Magnetization (h=0.02)
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

#Susceptibility
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
```

```python
plt.figure(figsize=(20,20))

plt.suptitle('1D Ising model',fontsize=20)
plt.subplot(221)
plt.title('Internal energy (per particle)')
bl, E, errE = np.loadtxt("Temp_var/output_ene_temp.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
blg, Eg, errEg = np.loadtxt("Temp_var/output_ene_tempG.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(bl, E ,yerr=errE, fmt = 'o', label = 'Metropolis')
plt.errorbar(blg, Eg ,yerr=errEg, fmt = '*', label = 'Gibbs')
plt.plot(T, e, label = 'Expected')
plt.xlabel('$T$')
plt.ylabel('$U/N$')
plt.legend(fontsize = 'large')
plt.grid(True)

plt.subplot(222)
plt.title('Heat capacity')
bl, h, errh = np.loadtxt("Temp_var/output_heat_temp.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
blg, hg, errhg = np.loadtxt("Temp_var/output_heat_tempG.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(bl, h ,yerr=errh, fmt = 'o', label = 'Metropolis')
plt.errorbar(blg, hg ,yerr=errhg, fmt = '*', label = 'Gibbs')
plt.plot(T, heat, label = 'Expected')
plt.xlabel('$T$')
plt.ylabel('$C$')
plt.legend(fontsize = 'large')
plt.grid(True)

plt.subplot(223)
plt.title('Average magnetization ($h=0.02$)')
bl, m, errm = np.loadtxt("Temp_var/output_mag_temp.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
blg, mg, errmg = np.loadtxt("Temp_var/output_mag_tempG.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(bl, m ,yerr=errm, fmt = 'o', label = 'Metropolis')
plt.errorbar(blg, mg ,yerr=errmg, fmt = '*', label = 'Gibbs')
plt.plot(T, M, label = 'Expected')
plt.xlabel('$T$')
plt.ylabel('$M$')
plt.legend(fontsize = 'large')
plt.grid(True)

plt.subplot(224)
plt.title('Average susceptibility')
bl, x, errx = np.loadtxt("Temp_var/output_chi_temp.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
blg, xg, errxg = np.loadtxt("Temp_var/output_chi_tempG.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
plt.errorbar(bl, x ,yerr=errx, fmt = 'o', label = 'Metropolis')
plt.errorbar(blg, xg ,yerr=errxg, fmt = '*', label = 'Gibbs')
plt.plot(T, X, label = 'Expected')
plt.xlabel('$T$')
plt.ylabel('$\chi$')
plt.legend(fontsize = 'large')
plt.grid(True)

plt.show()
```
