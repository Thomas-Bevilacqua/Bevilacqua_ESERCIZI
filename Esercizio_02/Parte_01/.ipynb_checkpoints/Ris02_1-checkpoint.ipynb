### Parte 1
1. Estraggo $10^4$ numeri casuali distribuiti uniformemente nell'intervallo $[0,1]$ e valuto l'integranda.
2. Dalle formule viste a lezione, posso valutare l'integrale calcolando la 'media' della funzione e moltiplicandola per la lunghezza dell'intervallo. Dato che l'intervallo è lungo 1, sfrutto direttamente la funzione 'block_unc' per valutare la media dell'integrale col metodo dei blocks.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, i, err = np.loadtxt("uniform_int.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(x,i,yerr=err)
plt.xlabel('$\#\,throws$')
plt.ylabel('$\int \pi/2 \, \cos(\pi x/2)\, dx$')
plt.grid(True)

plt.show()
```
