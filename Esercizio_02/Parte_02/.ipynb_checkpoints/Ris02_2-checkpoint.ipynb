## Esercizio 02.2
### Parte 1
1. Mi metto nell'origine per partire e faccio un passo. Per decidere la direzione, estraggo uniformemente un numero casuale tra $[-3,3]$ e mi muovo. Il tutto viene ripetuto per $10^4$ e per ciascuna rieptizione calcolo la distanza dall'origine e la salvo in un vettore.
2. Calcolo la media e la deviazione standard e la salvo.
3. Ripeto il tutto fino ad arrivare a 100 passi.
In questo caso nei risultati si ha la distanza media dall'origine al variare del numero di passi con rappresentata anche la deviazione standard: non uso il metodo dei blocks o la deviazione standard della media.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

#funzione di fit
def f(x,n):
    return n * np.sqrt(x)

nsteps, r, err = np.loadtxt("Discrete_RW.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(nsteps, r, yerr=err)
plt.xlabel('$\#\,steps$')
plt.ylabel('$<r>$')
plt.title('Discrete random walk')

#faccio il fit: prima ricavo il parametro n
#e l'incertezza, poi disegno
n_ott, n_err = curve_fit(f, nsteps, r)
y_fit = f(nsteps, n_ott)
plt.plot(nsteps,y_fit)

print("Stima per n =")
print(n_ott)
print("Incertezza su n =")
print(np.sqrt(n_err))
plt.show()
```

### Parte 2
Come nel caso precedente, ripeto da 1 a 100 passi per $10^4$ volte partendo dall'origine. Questa volta però uso coordinate sferiche per decidere il movimento: estraggo l'angolo solido con una distribuzione uniforme come indicato nel testo dell'esercizio.
Infine calcolo la media e la deviazione standard della distanza dall'origine al variare del numero di passi.

```python
nsteps, r, err = np.loadtxt("Continuum_RW.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(nsteps, r, yerr=err)
plt.xlabel('$\#\,steps$')
plt.ylabel('$<r>$')
plt.title('Continuum random walk')

#ripeto il fit come prima
n_ott, n_err = curve_fit(f, nsteps, r)
y_fit = f(nsteps, n_ott)
plt.plot(nsteps,y_fit)

print("Stima per n =")
print(n_ott)
print("Incertezza su n =")
print(np.sqrt(n_err))

plt.show()
```

I risultati sono in ottimo accordo con la funzione di fit: il comportamente è diffusivo (?)
