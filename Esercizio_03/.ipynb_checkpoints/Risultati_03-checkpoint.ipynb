```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt
```

Per calcolare il *put-option price* e il *call-option price* ho implementato le due relazioni per calcolare l'*asset price* indicate nel testo utilizzando i parametri assegnati; in particolare, per il sampling diretto ho usato
$$ S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right] $$
dove
- $S(0)=100$ è l'*asset price* al tempo iniziale $t=0$,
- $\mu$ coincide con l'*interest rate* $r=0.1$,
- $\sigma=0.25$ è la volatilità e 
- $W(t)$ è un numero random estratto secondo una distribuzione normale centrata in zero e con larghezza $t$.
Per quanto riguarda invece il *sampling discreto* ho utilizzato
$$ S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)\Delta t +\sigma Z_{i+1}\sqrt{\Delta t}\right] $$
dove ho usato un passo costante $\Delta t=\frac{T-t}{N_{int}}$ e $Z_i$ sono estratti secondo una distribuzione normale centrata in zero e con larghezza 1.

Dato dunque $S(t)$ ho calcolato il *call-option price* e il *put-option price* utilizzando queste due formule viste a lezione
$$ C[S(T),T]=e^{-rT}\max\left(0, S(T)-K\right) \\
P[S(T),T]=e^{-rT}\max\left(0, K-S(T)\right) $$
dove $K=100$ è lo *strike price* e $T=1$ è il *delivery time*.

Calcolo ora i valori attesi utilizzando la soluzione analitica presentata nel testo:

```python
# Data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t = 0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" -- BLACK-SCHOLES ANALYTIC SOLUTION (at time T=1) --")
    print("Put-option price =", put)
    print("Call-option price =", call)
```

Per la simulazione invece ho fatto $M=10^5$ valutazioni delle *options* e poi ne ho calcolato valor medio ed errore utilizzando il *blocking method*; nel caso discreto ho diviso l'intervallo temporale in 100 steps. Questi sono i risultati:

```python
plt.rc('text', usetex=True)

plt.figure(figsize=(15,6))
plt.suptitle('Direct sampling', fontsize = 20)

plt.subplot(121)
plt.title('Put-option price')
bl, Pd, errpd = np.loadtxt("Risultati/Put.dir", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, Pd, yerr=errpd, label='MC solution')
plt.axhline(y=put, color='red', linewidth=1, label=' B-S analytic solution')

plt.xlabel('M')
plt.ylabel('$P(S(T), T)$')
plt.grid(True)
plt.legend()

plt.subplot(122)
plt.title('Call-option price')
bl, Cd, errcd = np.loadtxt("Risultati/Call.dir", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, Cd, yerr=errcd, label='MC solution')
plt.axhline(y=call, color='red', linewidth=1, label=' B-S analytic solution')

plt.xlabel('M')
plt.ylabel('$C(S(T), T$')
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.suptitle('Discrete sampling', fontsize = 20)

plt.subplot(121)
plt.title('Put-option price')
bl, Pc, errpc = np.loadtxt("Risultati/Put.dis", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, Pc, yerr=errpc, label='MC solution')
plt.axhline(y=put, color='red', linewidth=1, label=' B-S analytic solution')

plt.xlabel('M')
plt.ylabel('$P(S(T), T)$')
plt.grid(True)
plt.legend(loc=4)

plt.subplot(122)
plt.title('Call-option price')
bl, Cc, errcc = np.loadtxt("Risultati/Call.dis", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, Cc, yerr=errcc, label='MC solution')
plt.axhline(y=call, color='red', linewidth=1, label=' B-S analytic solution')

plt.xlabel('M')
plt.ylabel('$C(S(T), T$')
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(15,12))
plt.suptitle('Direct sampling vs discrete sampling', fontsize = 20)

plt.subplot(211)
plt.title('Put-option price')
plt.errorbar(bl, Pd, yerr=errpd, label='Direct sampling')
plt.errorbar(bl, Pc, yerr=errpc, label='Dscrete sampling')
plt.axhline(y=put, color='red', linewidth=1, label=' B-S analytic solution')

plt.xlabel('M')
plt.ylabel('$P(S(T), T)$')
plt.grid(True)
plt.legend()

plt.subplot(212)
plt.title('Call-option price')
plt.errorbar(bl, Cd, yerr=errcd, label='Direct sampling')
plt.errorbar(bl, Cc, yerr=errcc, label='Discrete sampling')
plt.axhline(y=call, color='red', linewidth=1, label=' B-S analytic solution')

plt.xlabel('M')
plt.ylabel('$C(S(T), T$')
plt.grid(True)
plt.legend()

plt.show()
```

Dai grafici possiamo vedere che per questa scelta dei parametri i due metodi sono molto simili sia per quanto riguarda gli errori, sia per la convergenza verso il valore dato dalla soluzione analitica.
