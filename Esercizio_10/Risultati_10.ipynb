### <font color="blue"> Parte 01

Per risolvere questo esercizio ho aggiunto la funzione di *simulated annealing* al codice dell'esercizio precedente, per il resto il funzionamento è uguale. Questa funzione prende in input due configurazioni e valuta quale è la migliore tramite l'algoritmo Metropolis.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Circle(x, y, r):
    return x**2 + y**2 - r

def L2(xi, yi, xf, yf):
    return (xf - xi)**2 + (yf - yi)**2
```

```python
N = 100
r = 1.

# Construct the circle
xx = np.linspace(-r, r, N)
yy = np.linspace(-r, r, N)
X, Y = np.meshgrid(xx, yy)

plt.rc('text', usetex=True)
plt.figure(figsize=(8, 8))

plt.title('Cities disposition')
x, y = np.loadtxt("Parte_01/Risultati/circ.0", usecols=(0,1), unpack='true')
plt.plot(x, y, linestyle=' ', marker='.', label='Cities')
plt.plot(x[0], y[0], linestyle=' ', marker='.', color='red', label='Hometown')
plt.contour(X, Y, Circle(X, Y, r), [0], linewidths=0.5, colors='black')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.legend()

plt.show()
```

Quella in figura è la disposizione di 30 città utilizzata. Sono le stesse dell'esercizio precedente, dunque mi aspetto un percorso uguale o comunque simile.

Per la simulazione ho scelto i seguenti parametri:
- numero di mutazioni: 14,
- $\beta\in[0:70]$ con step 0.0005,
- MC steps: 1000.

Per ogni temperatura fissata avvengono 1000 MC steps, in cui ogni volta viene proposta una nuova disposizione che viene accettata tramite l'algoritmo di *simulated annealing*. L'ultima di queste viene poi salvata. I seguenti grafici sono costruiti in funzione della temperatura.

```python
# Compute the optimized path length
xc, yc = np.loadtxt("Parte_01/Risultati/circ.fin", usecols=(0,1), unpack='true')
dim = np.size(xc)
sum = 0.

for i in range(dim - 1):   # the last city is the first
    sum += L2(xc[i], yc[i], xc[i+1], yc[i+1])

print('Optimized configuration length = ', sum)
```

```python
plt.figure(figsize=(16,8))
plt.suptitle('Cities on a circle',fontsize=20)

plt.subplot(121)
plt.title('Best path length')
x, y= np.loadtxt("Parte_01/Risultati/circ.shortest", usecols=(0,1), unpack='true')
plt.plot(x, y)

plt.xlabel(r'$\beta$')
plt.ylabel('$L^{(2)}_{best}$')
plt.grid(True)

plt.subplot(122)
plt.title('Shortest path')
plt.plot(xc, yc, marker='.')
plt.plot(xc[0], yc[0], linestyle=' ', marker='.', color='red', label='Hometown')
plt.contour(X, Y, Circle(X, Y, r), [0], linewidths=0.5, colors='black')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.legend()

plt.show()
```

Osservo che nell'altro esercizio il percorso trovato era migliore dato che era più corto: in questo caso si può vedere (dai risultati numerici) che un paio di città molto vicine non vengono scambiate correttamente e dunque il percorso risulta un pochino più lungo.

### <font color="blue"> Parte 02
    
*Nota*: in quest'ultima parte ho scritto i codici relativi alla circonferenza e al quadrato in due file diversi, rispettivamente `circle.cpp` e `square.cpp`. Tuttavia è possibile compilare ed eseguire tutto insieme tramite il comando `make all` e poi `make run`.

Infine in questa seconda parte ho parallelizzato l'algoritmo: l'unica variazione del codice rispetto alla parte precedente consiste nelle righe necessarie al corretto funzionamento delle librerie *MPI*. Ho usato quattro nodi e ho cambiato lo step di $\beta$ fissandolo a 0.005.

I risultati mostrano il miglior percorso ottenuto (scelto tra i nodi) e la rispettiva lunghezza: ciascun nodo naturalmente utilizzava diversi numeri random.

```python
# Compute the optimized path length
xc, yc = np.loadtxt("Parte_02/Risultati/Circle.ris", usecols=(0,1), unpack='true')
xs, ys = np.loadtxt("Parte_02/Risultati/Square.ris", usecols=(0,1), unpack='true')
dim = np.size(xb)
sumc = 0.
sums = 0.

for i in range(dim - 1):   # the last city is the first
    sumc += L2(xc[i], yc[i], xc[i+1], yc[i+1])
    sums += L2(xs[i], ys[i], xs[i+1], ys[i+1])

print('Circle: optimized configuration path length = ', sumc)
print('Square: optimized configuration path length = ', sums)
```

```python
plt.figure(figsize=(16,8))
plt.suptitle('Parallel SA: shortest paths',fontsize=20)

plt.subplot(121)
plt.title('Circle')
plt.plot(xc, yc, marker='.')
plt.plot(xc[0], yc[0], linestyle=' ', marker='.', color='red', label='Hometown')
plt.contour(X, Y, Circle(X, Y, r), [0], linewidths=0.5, colors='black')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.grid(True)
plt.legend()

plt.subplot(122)
plt.title('Square')
plt.plot(xs, ys, marker='.')
plt.plot(xs[0], ys[0], linestyle=' ', marker='.', color='red', label='Hometown')
plt.axhline(y=-l, xmin=-l, xmax=l, color='black', linewidth=0.5)
plt.axhline(y=l, xmin=-l, xmax=l, color='black', linewidth=0.5)
plt.axvline(x=-l, ymin=-l, ymax=l, color='black', linewidth=0.5)
plt.axvline(x=l, ymin=-l, ymax=l, color='black', linewidth=0.5)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.grid(True)
plt.legend()

plt.show()
```

Osservo in conclusione che anche utilizzando un'esecuzione in parallelo del codice i risultati sono coerenti con quelli ottenuti negli altri modi: per quanto riguarda le città nel quadrato non si hanno differenze, mentre nel caso della circonferenza l'algoritmo genetico dell'esercizio precedente funziona meglio, dato che trova un percorso migliore.
