## Esercizio 10

Per risolvere questo esercizio ho aggiunto la funzione di simulated annealing al codice dell'esercizio precedente, per il resto il funzionamento è uguale. Questa funzione prende in input due configurazioni e valuta quale è la migliore tramite l'algoritmo Metropolis.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Circle(x, y, r):
    return x**2 + y**2 - r

def L2(xi, yi, xf, yf):
    return (xf - xi)**2 + (yf - yi)**2
```

```python
N = 100
r = 1.

# Construct the circle
xx = np.linspace(-r, r, N)
yy = np.linspace(-r, r, N)
X, Y = np.meshgrid(xx, yy)

plt.rc('text', usetex=True)
plt.figure(figsize=(8, 8))

plt.title('Cities disposition')
x, y = np.loadtxt("circ.0", usecols=(0,1), unpack='true')
plt.plot(x, y, linestyle=' ', marker='.', label='Cities')
plt.plot(x[0], y[0], linestyle=' ', marker='.', color='red', label='Hometown')
plt.contour(X, Y, Circle(X, Y, r), [0], linewidths=0.5, colors='black')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.grid(True)
plt.legend()

plt.show()
```

Quella in figura è la disposizione di 30 città utilizzata. Sono le stesse dell'esercizio precedente, dunque mi aspetto un percorso uguale.

Per la simulazione ho scelto i seguenti parametri:
- numero di mutazioni: 14,
- $\beta\in[0:70]$ with step 0.0005,
- MC steps: 1000.

Per ogni temperatura fissata avvengono 1000 MC steps e l'ultima configurazione viene salvata. I seguenti grafici sono costruiti in funzione della temperatura.

```python
# Compute the optimized path length
xc, yc = np.loadtxt("circ.fin", usecols=(0,1), unpack='true')
dim = np.size(xc)
sum = 0.

for i in range(dim - 1):   # the last city is the first
    sum += L2(xc[i], yc[i], xc[i+1], yc[i+1])

print('Optimized configuration length = ', sum)
```

```python
plt.figure(figsize=(16,8))
plt.suptitle('Cities on a circle',fontsize=20)

plt.subplot(121)
plt.title('Best path length')
x, y= np.loadtxt("circ.shortest", usecols=(0,1), unpack='true')
plt.plot(x, y, label='Data')
plt.axhline(y=sum, color='red', linewidth=0.7, label='Expected')

plt.xlabel('beta')
plt.ylabel('$L^{(2)}_{best}$')
plt.grid(True)
plt.legend()

plt.subplot(122)
plt.title('Shortest path')
plt.plot(xc, yc, marker='.')
plt.plot(xc[0], yc[0], linestyle=' ', marker='.', color='red', label='Hometown')
plt.contour(X, Y, Circle(X, Y, r), [0], linewidths=0.5, colors='black')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.grid(True)
plt.legend()

plt.show()
```

Osservo che nell'altro esercizio il percorso trovato era migliore dato che era più corto: in questo caso si può vedere (dai risultati numerici) che un paio di città non vengono scambiate correttaemente e dunque il percorso risulta un pochino più lungo.


Ripeto la simulazione con le città in un quadrato.

```python
l = 1.

plt.rc('text', usetex=True)
plt.figure(figsize=(8, 8))

plt.title('Cities disposition')
x, y = np.loadtxt("sq.0", usecols=(0,1), unpack='true')
plt.plot(x, y, linestyle=' ', marker='.', label='Cities')
plt.plot(x[0], y[0], linestyle=' ', marker='.', color='red', label='Hometown')

plt.axhline(y=-l, xmin=-l, xmax=l, color='black', linewidth=0.5)
plt.axhline(y=l, xmin=-l, xmax=l, color='black', linewidth=0.5)
plt.axvline(x=-l, ymin=-l, ymax=l, color='black', linewidth=0.5)
plt.axvline(x=l, ymin=-l, ymax=l, color='black', linewidth=0.5)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.grid(True)
plt.legend()

plt.show()
```

```python
# Compute the optimized path length
xb, yb = np.loadtxt("sq.fin", usecols=(0,1), unpack='true')
dim = np.size(xb)
sum = 0.

for i in range(dim - 1):   # the last city is the first
    sum += L2(xb[i], yb[i], xb[i+1], yb[i+1])

print('Optimized configuration path length = ', sum)
```

In questo caso invece noto che l'algoritmo di SA ha funzionato meglio rispetto all'altro esercizio.
