### Parte 02
Per minimizzare i parametri ho seguito questo procedimento:
- scelgo dei valori iniziali e finali per $\mu$ e $\sigma$, poi fisso uno step;
- costruisco un reticolo al variare di ogni possibile coppia e calcolo l'energia tramite l'algoritmo Metropolis per ogni coppia;
- al minimo dell'energia faccio corrispondere i valori preferiti di $\mu_{ott}$ e $\sigma_{ott}$.

Dopo qualche tentativo e osservando il potenziale, scelgo $\mu\in[0.7,1]$ in steps da 0.01 e $\sigma\in[0.4,0.8]$ in steps da 0.01. L'algoritmo Metropolis usa lo stesso numero di steps e di ampiezza di prima. 

Con questa scelta si ottiene $E_{min}=-0.453$, $\mu_{ott}=0.82$ e $\sigma_{ott}=0.6$. I risultati sono i seguenti:

```python
def psi2_T(x, mu, sigma):
    psi = np.exp( -(x-mu)**2/(2.*sigma**2) ) + np.exp( -(x+mu)**2/(2.*sigma**2) )
    norm = 2. * sigma * np.sqrt(np.pi) * ( 1. + np.exp(-mu**2/sigma**2) )
    return  psi**2/norm

def Vpot(x):
    return (x**2 - 2.5)*x**2
```

```python
bins = 100
N = 1000
mu = 0.82
sigma = 0.6

x = np.linspace(-2.5, 2.5, N)
dx = x[1] - x[0] # the step size

V = Vpot(x)
y = psi2_T(x, mu, sigma)

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-CDiff)/(2*dx**2) + np.diag(V)
E,psi = np.linalg.eigh(H)

psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

plt.figure(figsize=(20,5))
plt.suptitle('Results with optimized parameters',fontsize=20)

plt.subplot(121)
plt.title('Sampling of $|\psi(x)|^2$ (normalized)')
pr = np.loadtxt("Parte_01,02/Psi.min", usecols=(0), unpack='true')
n, bins, patches = plt.hist(pr, bins, range=(-2.5,2.5), density=True, label='Data')
plt.plot(x,y, label='Expected')
plt.plot(x,(psi[0])**2, label='Numerical')
plt.xlabel('$x$')
plt.ylabel('$\#$ of throws/$\#$ total throws')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.title('$<H>$')
x, y, erry = np.loadtxt("Parte_01,02/En.min", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(x, y, yerr=erry)
plt.xlabel('$\#$ of steps')
plt.ylabel('$<H>$')
plt.grid(True)

plt.show()
```
