## ESERCIZIO 01.1
### Parte 1
In questo esercizio ho replicato il codice in Python dato nella consegna. Ho implementato in particolare una funzione 'block_unc' per valutare l'errore statistico col metodo dei blocks.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, r, err = np.loadtxt("r_med.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(x,r-0.5,yerr=err)
plt.xlabel('$\#\,throws$')
plt.ylabel('$<r> - 0.5$')
plt.grid(True)

plt.show()
```

### Parte 2
Anche in questo secondo esercizio ho replicato il codice riportato nella consegna.

```python
x, sig, err = np.loadtxt("sigma.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(x,sig-1/12,yerr=err)
plt.xlabel('$ \#\,throws $')
plt.ylabel('$<r-0.5>^2 - 1/12$')
plt.grid(True)

plt.show()
```

### Parte 3
1. Genero $10^6$ numeri casuali distribuiti uniformemente in $[0,1]$ e li divido in 100 blocchi. Divido anche l'intervallo in 100 steps.
2. Fissato un blocco, scelgo il primo step: se un numero cade in questo step, incremento il contatore.
3. Incremento il valore di $\chi^2$ utilizzando la formula data nel testo: $n_i$ è il totale dei numeri caduti nello step.
4. Ripeto per ogni step fino a coprire l'intervallo: ho quindi una stima di $\chi^2$ per il primo blocco di numeri.
5. Ripeto per ogni blocco di numeri ottenendo 100 stime di $\chi^2$.

```python
x, chi = np.loadtxt("Chi2.dat", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x, chi)
plt.xlabel('$block\,(10^4\,random\,numbers)$')
plt.ylabel('$\chi^2$')
plt.grid(True)

plt.show()
```
