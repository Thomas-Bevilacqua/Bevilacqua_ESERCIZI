```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.stats import cauchy
```

### <font color="blue"> Parte 01

In questo esercizio ho calcolato gli integrali proposti utilizzando fondamentalmente il codice riportato nel notebook del testo.

Per calcolare l'incertezza ho implementato il *blocking method* nel file `func.cpp`. Questa funzione prende come input un vettore contenente i dati, il numero di blocks e una stringa col nome del file in cui salvare il risultato; quest'ultimo sar√† diviso in tre colonne che contengono il numero di dati utilizzati e il corrispondente valor medio con la relativa incertezza.

Ho utilizzato $10^6$ throws divisi in 100 blocks.

I risultati dei primi due integrali
$$ <r>=\int_0^1 dr~r = \frac{1}{2}, \\
\sigma^2=\int_0^1 dr~\left(r-\frac{1}{2}\right)^2=\frac{1}{12} $$
sono i seguenti:

```python
plt.rc('text', usetex=True)

plt.figure(figsize=(12,10))

plt.subplot(211)
x, r, err = np.loadtxt("Parte_01/Risultati/r.ave", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.axhline(y=0.5, color='red', linewidth=1, label='Expected')
plt.errorbar(x, r, yerr=err, label='MC estimation')

plt.xlabel('$\#$ of throws')
plt.ylabel('$<r>$')
plt.grid(True)
plt.legend()

plt.subplot(212)
x, sig, err = np.loadtxt("Parte_01/Risultati/r.sigma", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.axhline(y=1./12., color='red', linewidth=1, label='Expected')
plt.errorbar(x, sig ,yerr=err, label='MC estimation')

plt.xlabel('$\#$ of throws')
plt.ylabel('$\sigma^2$')
plt.grid(True)
plt.legend()

plt.show()
```

Si nota una convergenza al valore atteso (valutazione analitica dell'integrale) con una decrescita dell'errore; questa in particolare deve essere del tipo $\propto 1/\sqrt{N}$ dove $N$ √® il numero di throws.

Infine ho eseguito il test del $\chi^2$ con i parametri proposti nel testo, in particolare ho estratto $10^6$ numeri random uniformemente distribuiti in $[0,1)$ e li ho divisi in gruppi da $n=10^4$ (cio√® 100 blocks). Fissando poi $M=100$ (numero di sottointervalli in cui ho diviso $[0,1)$) mi aspetto $\chi^2\simeq100$. Il procedimento seguito √® il seguente:

1. Fissato un blocco $j$, scelgo il primo step $M$: se un numero cade in questo step, incremento il contatore.
2. Incremento il valore di $\chi^2$ utilizzando la formula data nel testo: $n_i$ √® il totale dei numeri caduti nello step.
3. Ripeto per ogni step fino a coprire l'intervallo: ho quindi una stima di $\chi^2_j$ per il blocco j-esimo di numeri.
4. Ripeto per ogni blocco di numeri ottenendo 100 stime di $\chi^2$.

Il risultato √® questo:

```python
plt.figure(figsize=(10,6))

x, chi = np.loadtxt("Parte_01/Risultati/Chi2.test", usecols=(0,1), delimiter='   ', unpack='true')
plt.plot(x, chi, label='$\chi^2$ estimated')
plt.axhline(y=100, color='red', linewidth=1, label='$\chi^2$ expected')

plt.xlabel('j-th block')
plt.ylabel('$\chi^2$')
plt.grid(True)
plt.legend()

plt.show()
```

Si nota un'oscillazione del $\chi^2$ intorno al valore atteso pari a 100 senza discostamenti troppo elevati: questo mostra che il generatore √® buono.

### <font color="blue"> Parte 02

Per prima cosa nel generatore di numeri casuali implemento le funzioni richieste per generare numeri random secondo le distribuzioni esponenziale e lorentziana.

Successivamente calcolo la funzione $S_N$ (media della somma di $N$ dadi) richiesta al variare di $N\in\left\lbrace 1,2,10,100 \right\rbrace$ per tutte le distribuzioni utilizzando i parametri forniti nel testo. Sottolineo che nel caso delle distribuzioni esponenziale e lorentziana ho estratto anche i numeri reali (e nel secondo caso negativi) e non solo gli interi da 1 a 6 come per il dado classico.

Ottengo i seguenti istogrammi che rappresentano il numero di eventi corrispondenti ad ogni valore di $S_N$; l'ultimo √® sempre normalizzato in modo da poterlo fittare con una gaussiana o lorentziana.

```python
plt.figure(figsize=(15,12))
plt.suptitle('Standard dice', fontsize = 20)

plt.subplot(221)
plt.title('N = 1')
x = np.loadtxt("Parte_02/Risultati/stand_1.dice", unpack='true')
plt.hist(x, 60)

plt.xlabel('$S_1$')
plt.ylabel('$\#$ events')
plt.grid(True)

plt.subplot(222)
plt.title('N = 2')
x = np.loadtxt("Parte_02/Risultati/stand_2.dice", unpack='true')
plt.hist(x, 60)

plt.xlabel('$S_2$')
plt.ylabel('$\#$ events')
plt.grid(True)

plt.subplot(223)
plt.title('N = 10')
x = np.loadtxt("Parte_02/Risultati/stand_10.dice", unpack='true')
plt.hist(x, 39)

plt.xlabel('$S_{10}$')
plt.ylabel('$\#$ events')
plt.grid(True)

plt.subplot(224)
plt.title('N = 100')
x = np.loadtxt("Parte_02/Risultati/stand_100.dice", unpack='true')
plt.hist(x, 49, density=True, label='Data')

# Fit
mu, sigma = norm.fit(x)
xmin, xmax = plt.xlim()
xx = np.linspace(xmin, xmax, 100)
y_fit = norm.pdf(xx, mu, sigma)

print('Optimized fit parameters: mean =', round(mu, 3), ', sigma^2 =', round(sigma**2, 3))
plt.plot(xx, y_fit, linewidth=1, color='red', label='Fit')

plt.xlabel('$S_{100}$')
plt.ylabel('$\#$ events (normalized)')
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(15,12))
plt.suptitle('Exponential dice', fontsize = 20)

plt.subplot(221)
plt.title('N = 1')
x = np.loadtxt("Parte_02/Risultati/exp_1.dice", unpack='true')
plt.hist(x, 60)

plt.xlabel('$S_1$')
plt.ylabel('$\#$ events')
plt.grid(True)

plt.subplot(222)
plt.title('N = 2')
x = np.loadtxt("Parte_02/Risultati/exp_2.dice", unpack='true')
plt.hist(x, 60)

plt.xlabel('$S_2$')
plt.ylabel('$\#$ events')
plt.grid(True)

plt.subplot(223)
plt.title('N = 10')
x = np.loadtxt("Parte_02/Risultati/exp_10.dice", unpack='true')
plt.hist(x, 60)

plt.xlabel('$S_{10}$')
plt.ylabel('$\#$ events')
plt.grid(True)

plt.subplot(224)
plt.title('N = 100')
x = np.loadtxt("Parte_02/Risultati/exp_100.dice", unpack='true')
plt.hist(x, 71, density=True, label='Data')

# Fit
mu, sigma = norm.fit(x)
xmin, xmax = plt.xlim()
xx = np.linspace(xmin, xmax, 100)
y_fit = norm.pdf(xx, mu, sigma)

print('Optimized fit parameters: mean =', round(mu, 3), ', sigma^2 =', round(sigma**2,3))
plt.plot(xx, y_fit, linewidth=1, color='red', label='Fit')

plt.xlabel('$S_{100}$')
plt.ylabel('$\#$ events (normalized)')
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(15,12))
plt.suptitle('Lorentzian dice', fontsize = 20)

plt.subplot(221)
plt.title('N = 1')
x = np.loadtxt("Parte_02/Risultati/lorentz_1.dice", unpack='true')
plt.hist(x, 60, range=(-60,60))

plt.xlabel('$S_1$')
plt.ylabel('$\#$ events')
plt.grid(True)

plt.subplot(222)
plt.title('N = 2')
x = np.loadtxt("Parte_02/Risultati/lorentz_2.dice", unpack='true')
plt.hist(x, 60, range=(-60,60))

plt.xlabel('$S_2$')
plt.ylabel('$\#$ events')
plt.grid(True)

plt.subplot(223)
plt.title('N = 10')
x = np.loadtxt("Parte_02/Risultati/lorentz_10.dice", unpack='true')
plt.hist(x, 60, range=(-60,60))

plt.xlabel('$S_{10}$')
plt.ylabel('$\#$ events')
plt.grid(True)

plt.subplot(224)
plt.title('N = 100')
x = np.loadtxt("Parte_02/Risultati/lorentz_100.dice", unpack='true')
plt.hist(x, 60, range=(-60,60), density=True, label='Data')

# Fit
mu, gamma = cauchy.fit(x)
xx = np.linspace(-20, 20, 100)
y_fit = cauchy.pdf(xx, mu, gamma)

print('Optimized fit parameters: mean =', round(mu, 3), ', Gamma =', round(gamma,3))
plt.plot(xx, y_fit, linewidth=1, color='red', label='Fit')

plt.xlabel('$S_{100}$')
plt.ylabel('$\#$ events (normalized)')
plt.grid(True)
plt.legend()

plt.show()
```

Noto che al crescere di $N$ i dati approssimano sempre meglio una distribuzione normale (o lorentziana) con media e varianza corrispondenti ai valori inseriti per creare i dati: il Teorema del Limite Centrale √® verificato dato che abbiamo considerato variabili generate con la stessa distribuzione e indipendenti tra loro.

### <font color="blue"> Parte 03

Per prima cosa scrivo una funzione nel file `func.cpp` per calcolare il seno di un angolo: potrei simulare un lancio generando un angolo in $[0,\pi]$ ma in tal caso dovrei utilizzare $\pi$ che invece voglio ricavare. Quindi procedo generando le coordinate $(x,y)$ nel semipiano superiore e calcolo il seno dell'angolo come
$$ \sin\theta=\frac{y}{\sqrt{x^2+y^2}}. $$

Utilizzo la lunghezza della sbarra $L=1$ e la distanza tra le linee $d=1.5$. Poi procedo cos√¨:
1. per simulare il lancio della sbarra estraggo la posizione del suo centro $l$ tra $[0,d]$ (distanza da una linea del pavimento) e due coordinate per valutarne l'angolo di caduta (rispetto alla linea del pavimento) con una distribuzione uniforme.
2. Successivamente, se la proiezione della sbarra sulla direzione perpendicolare alle linee del pavimento supera $l$ o √® pi√π piccola di $d-l$ allora vuol dire che interseca una riga del pavimento (o quella prima o quella dopo) e incremento un contatore.
3. Ripeto $10^4$ lanci e valuto $\pi$ con la formula del testo per $10^4$ volte. Dato che per estrarre in maniera uniforme un angolo tra $[0,\pi]$ le coordinate devono soddisfare
$$ x^2+y^2 < 1 $$
utilizzo un ciclo *while* e continuo finch√© non ho effettivamente $10^4$ lanci.
4. Infine utilizzo il *blocking method* per valutare l'errore (ho diviso i dati in 100 blocks).

Il valore di $\pi$ ottenuto √® il seguente:

```python
plt.figure(figsize=(10,6))

n, pi, err = np.loadtxt("Parte_03/Risultati/Pi.eval", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(n, pi, yerr=err, label='Buffon Experiment')
plt.axhline(y=np.pi, color='red', linewidth=1, label='Expected')

plt.xlabel('$\#$ of experiments')
plt.ylabel('$<\pi>$')
plt.grid(True)
plt.legend(loc=4)

plt.show()
```
