## ESERCIZIO 01.2
Le due parti sono realizzate insieme.
1. Modifico il file 'Random.cpp' e includo i due nuovi generatori di numeri casuali secondo le distribuzioni esponenziale e lorentziana.
2. Calcolo $S_N$ per $n=10^4$ volte per ciascuna distribuzione: dato che si tratta di dadi, genero i numeri casuali come 'double' e poi li tronco come 'int'; nel caso della lorentziana, dato che è centrata sullo 0 ed è simmetrica, prendo il modulo dei numeri generati.
3. In output ho quindi 4 files per ciascuna distribuzione al variare di $N$.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

#apro i dati
n_bins = 6
xu = np.loadtxt("stand_1.dat", usecols=(0), unpack='true')
xd = np.loadtxt("stand_2.dat", usecols=(0), unpack='true')
xdi = np.loadtxt("stand_10.dat", usecols=(0), unpack='true')
xc = np.loadtxt("stand_100.dat", usecols=(0), unpack='true')

#rappresento
n, bins, patches = plt.hist(xu, n_bins, range=(0,7))
plt.xlabel('$S_1$')
plt.ylabel('$\#\,events$')
plt.title('Fig1.1 - Dado standard, N = 1')
plt.grid(True)
plt.show()

n, bins, patches = plt.hist(xd, n_bins, range=(0,7))
plt.xlabel('$S_2$')
plt.ylabel('$\#\,events$')
plt.title('Fig1.2 - Dado standard, N = 2')
plt.grid(True)
plt.show()

n, bins, patches = plt.hist(xdi, n_bins, range=(0,7))
plt.xlabel('$S_{10}$')
plt.ylabel('$\#\,events$')
plt.title('Fig1.3 - Dado standard, N = 10')
plt.grid(True)
plt.show()

n, bins, patches = plt.hist(xc, n_bins, range=(0,7))
plt.xlabel('$S_{100}$')
plt.ylabel('$\#\,events$')
plt.title('Fig1.4 - Dado standard, N = 100')
plt.grid(True)
plt.show()
```
