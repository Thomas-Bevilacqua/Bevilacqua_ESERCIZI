## Molecular Dynamics
### Parte 1
Per quanto riguarda la possibilità di leggere un file con le vecchie posizioni ho modificato la funzione 'Input': ora funziona tramite un 'if' a seconda di ciò che viene specificato nel file 'input.dat', in particolare:
- se viene inserita la stringa 'actual' nell'ultima riga di 'input.dat' non vengono caricate le posizioni vecchie ma calcolate tramite le velocità (estratte con 'Random.cpp' in maniera uniforme tra $[-0.5,0.5]$).
- Se invece si inserisce 'old' vengono caricate direttamente le posizioni vecchie. Non vengono calcolate le velocità.

Successivamente, prima di iniziare la simulazione, viene raggiunta una temperatura di equilibrio $T^\star$ (da definire nel file di input) per il sistema grazie alla funzione 'Eq_Temp'. Il funzionamento è il seguente:
1. vengono effettuati 10 passi con l'algoritmo di Verlet per valutare la posizione a $t+dt$; ne scelgo 10 per avere una minima variazione dei parametri.
2. Viene calcolata la velocità e la temperatura dopo questi primi passi. Per calcolare la velocità ho modificato la funzione 'Move' in modo da salvare la posizione al tempo $t-dt$.
3. Con questa temperatura vengono riscalate le velocità.
4. Il tutto viene ripetuto finché la differenza tra due temperature consecutive raggiunge il valore di $\vert T^\star - T\vert=0.0001$.
5. Questa temperatura, queste posizioni e velocità vengono utilizzati come punto di partenza della simulazione.

### Parte 2
Per fare in modo che per ogni esecuzione nei file di output non vengano accumulati anche i risultati di simulazioni precedenti, ho modificato il 'Makefile' aggiungendo il comando './clean.sh' da eseguire prima della compilazione.

Per valutare le medie col metodo dei blocks, ho salvato su dei vettori i valori istantanei valutati con la funzione 'Measure'. Successivamente, tramite la funzione 'Measure_ave' vengono stampati su files i valori medi e le incertezze. Ho usato 100 blocks e la funzione 'block_unc'.

### Parte 3
Ho scritto la funzione 'Pressure' per misurare la pressione istantanea. Ogni volta che viene chiamata nella funzione 'Measure', valuta le forze agenti su ogni particella e le usa per calcolare il viriale. La pressione viene restituita direttamente dalla funzione.

Poi ho modificato le funzioni 'Measure' and 'Measure_ave' in modo da stampare su files i valori istantanei e medi della pressione.

### Parte 4 (Argon)
Eseguo la simulazione con i dati assegnati. I file di input utilizzati sono rinominati con 'input.gas', 'inpit.liquid', 'input.solid'; per ogni stato ho salvato i dati in cartelle e poi li ho riscalati in base al problema. Nel box sottostante vengono calcolate le grandezze utili al riscalamento.

Nei seguenti grafici si vedono solo 1000 step temporali perché le grandezze istantanee vengono valutate ogni 10 passi: per questo sono 1000 e non 10000. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy import constants

kb = constants.value('Boltzmann constant') #J/K
amu = constants.value('atomic mass constant') #kg-1

sig = 0.34*(10**-9) #m
e = 120. * kb #J
ek = 120. #e/kb
m = 39.948 * amu #kg
t = sig * np.sqrt(m/e)
pr = e/(sig**3)
```

```python
plt.rc('text', usetex=True)
plt.figure(figsize=(20,20))
plt.suptitle('Ag solid phase',fontsize=20)

plt.subplot(321)
plt.title('Average total energy')
bl, E, errE = np.loadtxt("Solid phase/ave_etot.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl * t,E * e,yerr=errE * e)
plt.xlabel('$t \, [s]$')
plt.ylabel('$<E_{tot}> \, [J]$')
plt.grid(True)

plt.subplot(322)
plt.title('Average kinetic energy')
bl, Ek, errEk = np.loadtxt("Solid phase/ave_ekin.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl * t,Ek * e,yerr=errEk * e)
plt.xlabel('$t \, [s]$')
plt.ylabel('$<E_k> \, [J]$')
plt.grid(True)

plt.subplot(323)
plt.title('Average potential energy')
bl, U, errU = np.loadtxt("Solid phase/ave_epot.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl * t,U * e,yerr=errU * e)
plt.xlabel('$t \, [s]$')
plt.ylabel('$<E_{pot}> \, [J]$')
plt.grid(True)

plt.subplot(324)
plt.title('Average temperature')
bl, T, errT = np.loadtxt("Solid phase/ave_temp.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl * t,T * ek,yerr=errT * ek)
plt.xlabel('$t \, [s]$')
plt.ylabel('$<T> \, [K]$')
plt.grid(True)

plt.subplot(325)
plt.title('Average pression')
bl, p, errp = np.loadtxt("Solid phase/ave_pres.dat", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl * t,p * pr,yerr=errp * pr)
plt.xlabel('$t \, [s]$')
plt.ylabel('$<p> \, [Pa]$')
plt.grid(True)

plt.show()
```
