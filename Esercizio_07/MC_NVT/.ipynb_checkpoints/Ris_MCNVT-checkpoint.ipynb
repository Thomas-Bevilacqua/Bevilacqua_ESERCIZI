Scrivo ora uno script per valutare la funzione di auto-correlazione $\chi(t)$ e la uso per rappresentare l'andamento dei dati appena prodotti. Il tempo è rappresentato dai MC steps e per chiarezza grafica decido di rappresentare solo l'intervallo $[0,100]$ (controllabile dal parametro 'rt').

Nel prossimo box riporto anche la funzione per il fit.

```python
def Auto_Correlation(t, obs):   # Auto-correlation function at time t
    t_max = len(obs)
    tp = t_max - t
    x_dev = np.sum(obs * obs)/t_max - (np.sum(obs)/t_max)**2   # st. deviation (denom.)
    
    x1 = obs[0:tp]     # x(t'): starts from t=0 to t'
    x2 = obs[t:t_max]  # x(t+t'): starts from t to the end
    
    return ( np.sum(x1 * x2)/tp - np.sum(x1) * np.sum(x2)/(tp**2) )/x_dev

def f(x, a, b):  # Fit function
    return b * np.exp(-x/a)
```

```python
rt = 100  # time range for plots
tt = np.linspace(0, rt, rt)

# Solid phase
chiE = np.zeros(rt)
chiP = np.zeros(rt)

Ei = np.loadtxt("Solid/output.epot.ist", usecols=(0), unpack='true')
Pi = np.loadtxt("Solid/output.pres.ist", usecols=(0), unpack='true')

for t in range(rt):
    chiE[t] = Auto_Correlation(t, Ei)
    chiP[t] = Auto_Correlation(t, Pi)

plt.figure(figsize=(20,5))
plt.suptitle('Solid phase',fontsize=20)

plt.subplot(121)
plt.plot(chiE, label='Data')
plt.title('Internal energy (per particle)')
plt.xlabel('t')
plt.ylabel('$\chi(t)$')
plt.yscale('Log')
plt.grid(True)

# Fit
tc_ott, tc_err = curve_fit(f, tt, chiE)
y_fit = f(tt, tc_ott[0], tc_ott[1])
plt.plot(tt, y_fit, label='Fit')
plt.legend()

print('SOLID PHASE:')
print('Energy correlation time $t_c$ =', tc_ott[0])
print('Error on energy $t_c$ =', tc_err[0][0])

plt.subplot(122)
plt.plot(chiP, label='Data')
plt.title('Pression')
plt.xlabel('t')
plt.ylabel('$\chi(t)$')
plt.yscale('Log')
#plt.xscale('Log')
plt.grid(True)

# Fit
tc_ott, tc_err = curve_fit(f, tt, chiP)
y_fit = f(tt, tc_ott[0], tc_ott[1])
plt.plot(tt, y_fit, label='Fit')
plt.legend()

print('Pression correlation time $t_c$ =', tc_ott[0])
print('Error on pression $t_c$ =', tc_err[0][0])
print('\n\n')

# Liquid phase
Ei = np.loadtxt("Liquid/output.epot.ist", usecols=(0), unpack='true')
Pi = np.loadtxt("Liquid/output.pres.ist", usecols=(0), unpack='true')

for t in range(rt):
    chiE[t] = Auto_Correlation(t, Ei)
    chiP[t] = Auto_Correlation(t, Pi)

plt.figure(figsize=(20,5))
plt.suptitle('Liquid phase',fontsize=20)

plt.subplot(121)
plt.plot(chiE, label='Data')
plt.title('Internal energy (per particle)')
plt.xlabel('t')
plt.ylabel('$\chi(t)$')
plt.yscale('Log')
plt.grid(True)

# Fit
tc_ott, tc_err = curve_fit(f, tt, chiE)
y_fit = f(tt, tc_ott[0], tc_ott[1])
plt.plot(tt, y_fit, label='Fit')
plt.legend()

print('LIQUID PHASE:')
print('Energy correlation time $t_c$ =', tc_ott[0])
print('Error on energy $t_c$ =', tc_err[0][0])

plt.subplot(122)
plt.plot(chiP, label='Data')
plt.title('Pression')
plt.xlabel('t')
plt.ylabel('$\chi(t)$')
plt.yscale('Log')
#plt.xscale('Log')
plt.grid(True)

# Fit
tc_ott, tc_err = curve_fit(f, tt, chiP)
y_fit = f(tt, tc_ott[0], tc_ott[1])
plt.plot(tt, y_fit, label='Fit')
plt.legend()

print('Pression correlation time $t_c$ =', tc_ott[0])
print('Error on pression $t_c$ =', tc_err[0][0])
print('\n\n')

# Gas phase
Ei = np.loadtxt("Gas/output.epot.ist", usecols=(0), unpack='true')
Pi = np.loadtxt("Gas/output.pres.ist", usecols=(0), unpack='true')

for t in range(rt):
    chiE[t] = Auto_Correlation(t, Ei)
    chiP[t] = Auto_Correlation(t, Pi)

plt.figure(figsize=(20,5))
plt.suptitle('Gas phase',fontsize=20)

plt.subplot(121)
plt.plot(chiE, label='Data')
plt.title('Internal energy (per particle)')
plt.xlabel('t')
plt.ylabel('$\chi(t)$')
plt.yscale('Log')
plt.grid(True)

# Fit
tc_ott, tc_err = curve_fit(f, tt, chiE)
y_fit = f(tt, tc_ott[0], tc_ott[1])
plt.plot(tt, y_fit, label='Fit')
plt.legend()

print('GAS PHASE:')
print('Energy correlation time $t_c$ =', tc_ott[0])
print('Error on energy $t_c$ =', tc_err[0][0])

plt.subplot(122)
plt.plot(chiP, label='Data')
plt.title('Pression')
plt.xlabel('t')
plt.ylabel('$\chi(t)$')
plt.yscale('Log')
#plt.xscale('Log')
plt.grid(True)

# Fit
tc_ott, tc_err = curve_fit(f, tt, chiP)
y_fit = f(tt, tc_ott[0], tc_ott[1])
plt.plot(tt, y_fit, label='Fit')
plt.legend()

print('Pression correlation time $t_c$ =', tc_ott[0])
print('Error on pression $t_c$ =', tc_err[0][0])

plt.show
```

Scrivo ora uno script per valutare l'incertezza al variare del numero di blocks e poi lo applico ai dati di prima. Dato che voglio rappresentare l'incertezza al variare della lunghezza dei blocks, decido di far restituire alla funzione l'errore dell'ultimo block corrispondente.

Campiono l'intervallo di lunghezza dei blocks $[10,5010]$ dividendolo in steps da 100.

```python
def dev_st(mean, mean2, n):
    if n==0:
        return 0
    else:
        return np.sqrt((mean2[n] - mean[n]**2)/n)
    
def Blocks_Err(L, obs):   # length of each block
    nsteps = len(obs)
    nbl = int(nsteps/L)
    
    mean = np.zeros(nbl) 
    mean2 = np.zeros(nbl)
    sum_prog = np.zeros(nbl)
    sum2_prog = np.zeros(nbl)
    err_prog = np.zeros(nbl)
        
    for i in range(nbl):
        s = 0.
        for j in range(L):
            s += obs[j + i*L]
        mean[i] = s/L            # mean for each block
        mean2[i] = (mean[i])**2
    
    for i in range(nbl):
        for j in range(i+1):
            sum_prog[i] += mean[j]
            sum2_prog[i] += mean2[j]
            
        sum_prog[i]/=(i+1)
        sum2_prog[i]/=(i+1)
        err_prog[i] = dev_st(sum_prog, sum2_prog, i) # Statistical uncertainty
    
    return err_prog[nbl-1]
```

```python
Li = 10
Lf = 5010   # for simplify the division
Lstep = 100
L = np.linspace(Li, Lf)
LL = int((Lf - Li)/Lstep)

# Solid phase
Ei = np.loadtxt("Solid/output.epot.ist", usecols=(0), unpack='true')
Pi = np.loadtxt("Solid/output.pres.ist", usecols=(0), unpack='true')
#print('Data loaded \n')

Eibl = np.zeros(LL) 
Pibl = np.zeros(LL)
Ls = np.zeros(LL)

i = 0
for l in range(Li, Lf, Lstep):
    #print('Current length =', l)
    Eibl[i] = Blocks_Err(l, Ei)
    Pibl[i] = Blocks_Err(l, Pi)
    Ls[i] = i * Lstep
    i+=1
    
plt.figure(figsize=(20,5))
plt.suptitle('Solid phase: statistical uncertaints with data blocking',fontsize=20)

plt.subplot(121)
plt.plot(Ls, Eibl)
plt.title('Internal energy (per particle)')
plt.xlabel('Blocks length')
plt.ylabel('$\sigma$')
plt.grid(True)

plt.subplot(122)
plt.plot(Ls, Pibl)
plt.title('Pression')
plt.xlabel('Blocks length')
plt.ylabel('$\sigma$')
plt.grid(True)

plt.show

# Liquid phase
Ei = np.loadtxt("Liquid/output.epot.ist", usecols=(0), unpack='true')
Pi = np.loadtxt("Liquid/output.pres.ist", usecols=(0), unpack='true')
#print('Data loaded \n')

Eibl = np.zeros(LL) 
Pibl = np.zeros(LL)
Ls = np.zeros(LL)

i = 0
for l in range(Li, Lf, Lstep):
    #print('Current length =', l)
    Eibl[i] = Blocks_Err(l, Ei)
    Pibl[i] = Blocks_Err(l, Pi)
    Ls[i] = i * Lstep
    i+=1
    
plt.figure(figsize=(20,5))
plt.suptitle('Liquid phase: statistical uncertaints with data blocking',fontsize=20)

plt.subplot(121)
plt.plot(Ls, Eibl)
plt.title('Internal energy (per particle)')
plt.xlabel('Blocks length')
plt.ylabel('$\sigma$')
plt.grid(True)

plt.subplot(122)
plt.plot(Ls, Pibl)
plt.title('Pression')
plt.xlabel('Blocks length')
plt.ylabel('$\sigma$')
plt.grid(True)

plt.show

# Gas phase
Ei = np.loadtxt("Gas/output.epot.ist", usecols=(0), unpack='true')
Pi = np.loadtxt("Gas/output.pres.ist", usecols=(0), unpack='true')
#print('Data loaded \n')

Eibl = np.zeros(LL) 
Pibl = np.zeros(LL)
Ls = np.zeros(LL)

i = 0
for l in range(Li, Lf, Lstep):
    #print('Current length =', l)
    Eibl[i] = Blocks_Err(l, Ei)
    Pibl[i] = Blocks_Err(l, Pi)
    Ls[i] = i * Lstep
    i+=1
    
plt.figure(figsize=(20,5))
plt.suptitle('Gas phase: statistical uncertaints with data blocking',fontsize=20)

plt.subplot(121)
plt.plot(Ls, Eibl)
plt.title('Internal energy (per particle)')
plt.xlabel('Blocks length')
plt.ylabel('$\sigma$')
plt.grid(True)

plt.subplot(122)
plt.plot(Ls, Pibl)
plt.title('Pression')
plt.xlabel('Blocks length')
plt.ylabel('$\sigma$')
plt.grid(True)

plt.show
```
