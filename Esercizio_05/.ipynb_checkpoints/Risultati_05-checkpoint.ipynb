Per prima cosa ho implementato le due funzioni d'onda: prendono in input un vettore con le coordinate cartesiane e restituiscono il modulo quadro della funzione d'onda.

Poi ho scritto l'algoritmo Metropolis per campionarle; come input questa funzione prende:
- la funzione da campionare,
- la posizione iniziale,
- il numero di steps da effettuare,
- il numero di steps da scartare all'inizio (per equilibrare il sistema),
- una stringa per indicare la probabilità di transizione (*Unif* per uniforme che estrae un numero random in $[-step,step]$ e *Gauss* per normale centrata in 0 e larga $step$),
- la larghezza dello step (cioè *step* del punto precedente),
- una stringa per indicare su che file stampare i punti campionati (coordinate cartesiane).

Il funzionamento dell'algoritmo è fondamentalmente il seguente:
1. per ogni direzione $(x,y,z)$ viene estratto il passo secondo la distribuzione uniforme o normale scelta;
2. viene effettuato il passo e viene calcolata la funzione da campionare nel nuovo punto;
3. viene calcolato il rapporto tra la funzione valutata nel punto nuovo e in quello vecchio e in base ai criteri visti a lezione il punto viene accettato o rigettato.

Il file contenente le coordinate viene poi caricato dalla funzione *Raggi* che calcola la distanza di ogni punto dall'origine.

Per prima cosa ho campionato le funzioni d'onda estraendo lo step in maniera uniforme. Ho usato $10^6$ throws, di cui ho scartato i primi 1000 e per calcolare le distanze medie ho utilizzato 100 blocks. Come punto di partenza ho scelto l'origine.

In modo da avere circa il 50% dei punti totali accettati ho scelto per la $\psi_{1,0,0}$ uno $step=1.2$ e per la $\psi_{2,1,0}$ uno $step=3$.

Sotto rappresento in ordine i punti campionati per i due orbitali proposti e il raggio medio (tutto in unità di $a_0$):

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.rc('text', usetex=True)

X, Y, Z = np.loadtxt("Risultati/psi_100.uniform", usecols=(0,1,2), delimiter='    ', unpack='true')
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_title('Ground state (1,0,0)')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.view_init(10, 30)
plt.show()

X, Y, Z = np.loadtxt("Risultati/psi_210.uniform", usecols=(0,1,2), delimiter='    ', unpack='true')
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_title('$2p$ excited state (2,1,0)')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.view_init(10, 30)

plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.suptitle('Average radius (uniform probability transition)',fontsize=20)

plt.subplot(121)
plt.title('Ground state $\psi_{1,0,0}$', fontsize = 16)
bl, rgu, errgu = np.loadtxt("Risultati/r100.uniform", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, rgu, yerr=errgu, label='Metropolis estimation')
plt.axhline(y=1.5, color='red', linewidth=1, label='Expected')

plt.xlabel('$\#$ of throws')
plt.ylabel('$<r>$')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.title('$2p$ state $\psi_{2,1,0}$', fontsize = 16)
bl, reu, erreu = np.loadtxt("Risultati/r210.uniform", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, reu, yerr=erreu, label='Metropolis estimation')
plt.axhline(y=5, color='red', linewidth=1, label='Expected')

plt.xlabel('$\#$ of throws')
plt.ylabel('$<r>$')
plt.legend()
plt.grid(True)

plt.show()
```

Successivamente provo a rifare la simulazione scegliendo come punto di partenza $(10,10,10)$, molto lontano dall'origine. Ottengo i seguenti risultati:

```python
plt.rc('text', usetex=True)

X, Y, Z = np.loadtxt("Risultati_L/psi_100.uniform", usecols=(0,1,2), delimiter='    ', unpack='true')
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_title('Ground state (1,0,0)')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.view_init(10, 30)
plt.show()

X, Y, Z = np.loadtxt("Risultati_L/psi_210.uniform", usecols=(0,1,2), delimiter='    ', unpack='true')
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_title('$2p$ excited state (2,1,0)')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.view_init(10, 30)

plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.suptitle('Average radius (uniform probability transition) starting far from the origin',fontsize=20)

plt.subplot(121)
plt.title('Ground state $\psi_{1,0,0}$', fontsize = 16)
bl, r, err = np.loadtxt("Risultati_L/r100.uniform", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, r, yerr=err, label='Metropolis estimation')
plt.axhline(y=1.5, color='red', linewidth=1, label='Expected')

plt.xlabel('$\#$ of throws')
plt.ylabel('$<r>$')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.title('$2p$ state $\psi_{2,1,0}$', fontsize = 16)
bl, r, err = np.loadtxt("Risultati_L/r210.uniform", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, r, yerr=err, label='Metropolis estimation')
plt.axhline(y=5, color='red', linewidth=1, label='Expected')

plt.xlabel('$\#$ of throws')
plt.ylabel('$<r>$')
plt.legend(loc=4)
plt.grid(True)

plt.show()
```

Anche in questo caso sono stati scartati i primi 1000 punti campionati ma non sono stati sufficienti per avere un risultato buono come nel caso precedente: conviene dunque partire dall'origine (o eventualmente molto vicino) oppure è necessario aumentare gli steps di equilibratura e il numero di steps.

Infine provo a campionare i punti utilizzando una probabilità di transizione normale: in questo caso ho cambiato la larghezza degli steps scegliendo per $\psi_{1,0,0}$ uno $step=0.8$ e per la $\psi_{2,1,0}$ uno $step=1.8$ in modo da avere circa il 50% dei punti accettati. Di nuovo ottengo i seguenti risultati:

```python
plt.rc('text', usetex=True)

X, Y, Z = np.loadtxt("Risultati/psi_100.norm", usecols=(0,1,2), delimiter='    ', unpack='true')
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_title('Ground state (1,0,0)')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.view_init(10, 30)
plt.show()

X, Y, Z = np.loadtxt("Risultati/psi_210.norm", usecols=(0,1,2), delimiter='    ', unpack='true')
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_title('$2p$ excited state (2,1,0)')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.view_init(10, 30)

plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.suptitle('Average radius (normal probability transition)',fontsize=20)

plt.subplot(121)
plt.title('Ground state $\psi_{1,0,0}$', fontsize = 16)
bl, rgg, errgg = np.loadtxt("Risultati/r100.norm", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, rgg, yerr=errgg, label='Metropolis estimation')
plt.axhline(y=1.5, color='red', linewidth=1, label='Expected')

plt.xlabel('$\#$ of throws')
plt.ylabel('$<r>$')
plt.legend(loc=4)
plt.grid(True)

plt.subplot(122)
plt.title('$2p$ state $\psi_{2,1,0}$', fontsize = 16)
bl, reg, erreg = np.loadtxt("Risultati/r210.norm", usecols=(0,1,2), delimiter='    ', unpack='true')
plt.errorbar(bl, reg, yerr=erreg, label='Metropolis estimation')
plt.axhline(y=5, color='red', linewidth=1, label='Expected')

plt.xlabel('$\#$ of throws')
plt.ylabel('$<r>$')
plt.legend()
plt.grid(True)

plt.show()
```

Infine confronto l'andamento dei due raggi medi:

```python
plt.figure(figsize=(15,12))
plt.suptitle('Average radius',fontsize=20)

plt.subplot(211)
plt.title('Ground state $\psi_{1,0,0}$', fontsize = 16)
plt.errorbar(bl, rgu, yerr=errgu, label='Uniform steps')
plt.errorbar(bl, rgg, yerr=errgg, label='Normal steps')
plt.axhline(y=1.5, color='red', linewidth=1, label='Expected')

plt.xlabel('$\#$ of throws')
plt.ylabel('$<r>$')
plt.legend()
plt.grid(True)

plt.subplot(212)
plt.title('$2p$ state $\psi_{2,1,0}$', fontsize = 16)
plt.errorbar(bl, reu, yerr=erreu, label='Uniform steps')
plt.errorbar(bl, reg, yerr=erreg, label='Normal steps')
plt.axhline(y=5, color='red', linewidth=1, label='Expected')

plt.xlabel('$\#$ of throws')
plt.ylabel('$<r>$')
plt.legend()
plt.grid(True)

plt.show()
```

Per quanto riguarda il valor medio dei raggi, si nota come i risultati ottenuti con entrambe le probabilità di transizione convergano verso il valore atteso con la stessa rapidità (nel senso che dopo circa lo stesso numero di steps i risultati sono simili e confrontabili col valore atteso) e che anche gli errori sono confrontabili, come mostrato sotto.

Inoltre dagli *scatter plots* precedenti non si notano particolari differenze nella distribuzione dei punti campionati.

```python
plt.figure(figsize=(15,6))
plt.suptitle('Relative error ($\%$)',fontsize=20)

plt.subplot(121)
plt.title('Ground state $\psi_{1,0,0}$')
plt.plot(bl, errgu/rgu*100, label='Uniform steps')
plt.plot(bl, errgg/rgg*100, label='Normal steps')

plt.xlabel('$\#$ of throws')
plt.ylabel(r'$\Delta r/r ~ \%$')
plt.grid(True)
plt.legend()

plt.subplot(122)
plt.title('$2p$ state $\psi_{2,1,0}$')
plt.plot(bl, erreu/reu*100, label='Uniform steps')
plt.plot(bl, erreg/reg*100, label='Normal steps')

plt.xlabel('$\#$ of throws')
plt.ylabel(r'$\Delta r/r ~ \%$')
plt.grid(True)
plt.legend()

plt.show()
```
